```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = fetch_california_housing()
df = pd.DataFrame(data["data"], columns=data["feature_names"])
df
```

```python
# Get categorical columns
cat_cols = df.select_dtypes(include=["object"]).columns.tolist()
cat_cols
```

```python
# Get continuous columns
cont_cols = df.select_dtypes(include=["number"]).columns.tolist()
cont_cols
```

```python
data_continuous = df[cont_cols]
data_continuous.head()
```

```python
summary_continuous = data_continuous.describe().transpose()
summary_continuous
```

```python
data_continuous.head()
```

```python
import numpy as np


def plot_numericals(data, cols):
    summary = data[cols].describe().transpose().sort_values(by="count")

    print(summary)

    n = data.shape[0]
    b = int(np.sqrt(n))
    for k, (col, val) in enumerate(summary["count"].items()):
        plt.figure()
        data[col].plot.hist(bins=b)
        plt.title(f"{col}, n={int(val)}")
        plt.show()


plot_numericals(data_continuous, data_continuous.select_dtypes("number").columns)
```

```python
# Clean up the data
data_continuous["MedInc"] = data_continuous["MedInc"].apply(np.log10)
```

```python
# fuction that check hom many nas are in the data
def check_nas(data):
    for col in data.columns:
        print(col, data[col].isna().sum())


check_nas(data_continuous)
```

```python
# function that compares col to target and plots it
def compare_to_target(data, target):
    for col in data.columns:
        if col == target.name:
            continue
        plt.figure()
        plt.scatter(data[col], target, alpha=0.1)
        plt.title(f"{col} vs target")
        plt.show()


compare_to_target(data_continuous, data_continuous["MedInc"])
```

```python
# Linear regression with one variable
# Target: MedInc
# Feature: AveRooms

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split


X = data_continuous["AveRooms"].values.reshape(-1, 1)
y = data_continuous["MedInc"].values.reshape(-1, 1)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

plt.figure()
plt.scatter(X_test, y_test, alpha=0.1)
plt.plot(X_test, y_pred, color="red")
plt.title("AveRooms vs MedInc")
plt.show()
```
